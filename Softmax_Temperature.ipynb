# Softmax Temperature Playground

The Softmax operation is a commonly used transformation from logits (predictions of a classifier) to a probability distribution, where the sum of all values is 1.

The Softmax operation can be enhanced by a temperature value that controls the "smoothness" of the resulting distribution, a technique often used in knowledge distilation or reinforcement learning. 

In this notebook we explore the effect of the temperature on the resulting distribution.

- Step 1: Simulate Data
- Step 2: Calculate Softmax
- Step 3: Interpretation

Softmax Equation:
$P(x)^{(i)} = \frac{e^{\frac{x_i}{\tau}}}{\sum_j e^{\frac{x_j}{\tau}}}$, 

where $x$ are the input logits, $i$ is the index of the i-th logit, and $\tau$ is the temperature with $\tau > 0$

```
import matplotlib.pyplot as plt
import numpy as np
```

```
def plot(labels, logits, title):
  fig, ax = plt.subplots()
  colors = ['#1b9e77', '#a9f971', '#fdaa48','#6890F0','#A890F0']
  ax.bar(labels, logits, color=colors)
  ax.set_title(title)

  plt.show()

def softmax(x, temp):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(np.divide(x,temp)) / np.sum(np.exp(np.divide(x,temp)), axis=0)
```

## Step 1: Simulate data
Simulate prediction scores of a classifier with 10 classes

```
logits = [
    0.4,
    0.3,
    0.1,
    0.02,
    0.38,
    0.18,
    0.35,
    0.44,
    0.22,
    0.59,
]
labels = [i for i in range(len(logits))]
```

```
plot(labels, logits, "Logits")
```

# Step 2: Calculate Softmax for different Temperature values.
Loop over different temperature values and calculate the softmax

```
temps = [0.1, 0.5, 1, 2, 5, 10, 100]
for temp in temps:
  sm = softmax(logits, temp)
  plot(labels, sm, f"Softmax, temperature: {temp}")
```

## Step 3: Interpretation

1. the higher the temperature, the more uniform is the probability distribution, meaning it also increases the randomnes if you would draw a sample from that distribution
2. the lower the temperature, the "sharper" the resulting distribution.
3. High temperatures boost low probable events and damps high probable ones
4. Low temperatures boost high probable events and damp low probable ones even more.
